EDA using Univariate Analysis

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.sample(5)
```

## 1. Categorical Data

### a. countplot

```python
# sns.countplot(df['Pclass'])
df['Survived'].value_counts().plot(kind= 'bar')
```

### b. piechart

```python
df['Sex'].value_counts().plot(kind='pie', autopct = '%0.2f')
```

## 2. Numerical Data

### a. Histogram
### b. Distplot
### c. Boxplot

```python
# Histogram
import matplotlib.pyplot as plt
plt.hist(df['Age'])
```

```python
#Distplot
sns.distplot(df['Age'])
```

```python
# Boxplot
# Boxplot gives 5 number summary
# minimum, maximum, median, Q1, Q2
```

```python
sns.boxplot(df['Age'], orient='h')
```
