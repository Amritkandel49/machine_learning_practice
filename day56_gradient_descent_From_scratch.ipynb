```python
import numpy as np
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, random_state=42, noise=80)
```

```python
X.shape, y.shape
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X, y)
```

```python
reg.coef_, reg.intercept_
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color = 'red')
```
