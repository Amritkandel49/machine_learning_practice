### Fetching data from Web scrapping

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.collegenp.com/colleges'
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get(url,headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'html')
```

```python
college_name = []
college_address = []
isFeatured = []
courses = []
for container in soup.find_all('div', class_="college-detail-content"):
    college_name.append(container.find('h3', class_="college-name").text)
    college_address.append(container.find('span', class_ = 'college-address').text)
    isFeatured.append(True if (container.find('div', class_ = 'featured-college')) else False)
    temp = []
    for uls in container.find_all('ul', "college-list-courses"):
        for lis in uls.find_all('li'):
            temp.append(lis.text)
        courses.append(temp)
isFeatured
```

```python
df = pd.DataFrame(
    {
        "name" :college_name,
        "address" : college_address,
        "courses available" : courses,
        "isFeatured" : isFeatured
    }
)
```
